A very simple environment example with random action.

```python
import gym
env = gym.make('FetchPickAndPlace-v1') # Change Swimmer-v2 for the other environment.
env.reset()
for _ in range(1000):
    env.render()
    env.step(env.action_space.sample()) # take a random action
env.close()
```

Simple environment with observation shown.

```python
env = gym.make('MountainCar-v0') # Change Swimmer-v2 for the other environment.
for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        env.render()
        print(observation)
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()
```
